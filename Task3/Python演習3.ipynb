### 問1

自然数を入力とし、入力された自然数が3の倍数(Ex. 3, 6, 9, ...)であるときは`hoge`、3を含む数(Ex. 3, 13, 23, ...)であるときは`huga`、両方の条件を満たす時は`piyo`を出力する関数`nabeatsu`を定義してください

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
6

# テストケース2
13

# テストケース3
33
```

````python
def nabeatsu(num:int) -> str:
    multiple = (num % 3 == 0)
    contains = ('3' in str(num))

    if multiple and contains:
        return 'piyo'
    elif multiple:
        return 'hoge'
    elif contains:
        return 'huga'
    else:
        return ''
    return
````

````python
case_1 = 6
case_2 = 13
case_3 = 33

print('case_1:', nabeatsu(case_1))
print('case_2:', nabeatsu(case_2))
print('case_3:', nabeatsu(case_3))
````

### 問2

異なる自然数を含む配列とターゲットとなる自然数を入力として、配列に含まれる自然数を足し合わせてターゲットとなる自然数になる組み合わせを出力する関数`combination_sum`を定義してください

このとき、同じ自然数は何度でも利用してよいものとし、もし組み合わせが存在しない場合は空の配列を返してください

例えば、配列`[2, 3, 6, 7]`、ターゲット`7`を入力としたときの出力は`[[2,2,3],[7]]`になります

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
配列: [2, 3, 6, 7]
ターゲット: 14

# テストケース2
配列: [2, 3, 5]
ターゲット: 8

# テストケース3
配列: [2, 9, 11]
ターゲット: 100
```

````python
from typing import List
def combination_sum(candidates:list, target:int) -> list:
    res = []          # 存放最终所有符合条件的组合
    candidates.sort()       # 排序，保证组合不重复（顺序统一）

    def process(start, path, rest):
        if rest == 0: #说明刚好满足，将其添加入结果集
            res.append(path[:])
            return

        # 如果剩余值小于 0，说明已经超出，直接剪枝
        if rest < 0:
            return

        # 从 start 开始选择，避免出现顺序不同但内容相同的组合
        for i in range(start, len(candidates)):
            path.append(candidates[i])               # 选择当前数字
            process(i, path, rest - candidates[i])       # 递归，i表示可重复使用当前数字
            path.pop()                          #回溯，撤销选择

    process(0, [], target)
    return res
````

解释一下发生了什么，首先是定义了一个 combination_sum 函数，接受候选数字列表 candidates 和目标值 target 作为参数。函数内部定义了一个递归函数 process，用于生成所有可能的组合。通过递归和回溯的方法，函数不断尝试添加候选数字到当前路径 path 中，并检查剩余值 rest 是否满足条件。当找到一个满足条件的组合时，将其添加到结果集 res 中。最终返回所有符合条件的组合列表 res。

很显然，这比上一道题目要复杂一些，但本质还是排列组合问题，要避免重复，然后设计好递归函数体，想明白了的话其实就不难

前の問題と比べて処理が複雑になっているが、本質的には排列・組み合わせの問題である。重複を避ける工夫を行い、再帰関数の構造を正しく設計できれば、理解自体はそれほど難しくない。

````python
candidates_1 = [2, 3, 6, 7]
target_1 = 14

candidates_2 = [2, 3, 5]
target_2 = 8

candidates_3 = [2, 9, 11]
target_3 = 100

print(combination_sum(candidates_1, target_1))
print(combination_sum(candidates_2, target_2))
print(combination_sum(candidates_3, target_3))
````

### 問3

括弧を含む文字列を入力として、括弧の内包関係が正しいかどうかを判定し、正しければ`True`、正しくなければ`False`を返す関数`format_checker`を定義してください

例えば、`[](){}`や`[({})]`は正しい、`[(]`や`{[(})]`は正しくないと判定します

また、以下のテストケースを入力した際の出力結果を示してください

```
# テストケース1
Hello I’m [(firstname) (lastname)] and I’m {age} years old!

# テストケース2
Hello I’m [(firstname) (lastname]) and I’m {age} years old!

# テストケース3
Hello I’m ((firstname) (lastname) and I’m {age} years old!
```

````python
def format_checker(text:str) -> bool:
    stack = []  # 用来存放左括号
    pair = {
        ')': '(',
        ']': '[',
        '}': '{'
    }

    for char in text:
        # 如果是左括号，直接入栈
        if char in '([{':
            stack.append(char)
        # 如果是右括号，检查是否和栈顶匹配
        elif char in ')]}':
            # 栈为空，说明没有可匹配的左括号
            if not stack:
                return False

            # 栈顶不是对应的左括号，结构错误
            if stack[-1] != pair[char]:
                return False

            # 匹配成功，弹出栈顶
            stack.pop()

        # 其他字符直接忽略

    # 字符串遍历完后，栈必须为空才算正确
    return len(stack) == 0
````

这就是最典型的stack问题啊，抓住重点，先 进 后 出

これは典型的なstack問題！懐かしい～

後入れ先出しという点を把握すれば簡単な問題に過ぎない


左括弧をstackに積み、右括弧が出現した時に
スタック最上位要素との対応を確認する。
最終的にstackが空であれば正しい。

````python
case_1 = 'Hello I’m [(firstname) (lastname)] and I’m {age} years old!'
case_2 = 'Hello I’m [(firstname) (lastname]) and I’m {age} years old!'
case_3 = 'Hello I’m ((firstname) (lastname) and I’m {age} years old!'

print('case_1:', format_checker(case_1))
print('case_2:', format_checker(case_2))
print('case_3:', format_checker(case_3))
````
