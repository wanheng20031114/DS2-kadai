```python
import requests
from bs4 import BeautifulSoup
import sqlite3
import time
```

```python
db_path = "google.db"
conn = sqlite3.connect(db_path)
cur = conn.cursor()
try:
    cur.execute("""
    CREATE TABLE IF NOT EXISTS repos (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT,
        language TEXT,
        stars INTEGER
    )
    """)
    conn.commit()
except sqlite3.Error as e:
    print(f"error : {e}")
finally:
    conn.close()
```

```python
headers = {"User-Agent": "Mozilla/5.0"}
base_url = "https://github.com/orgs/google/repositories?page="
total_pages = 94
db_path = "google.db"


all_repos = []

for page in range(1, total_pages + 1):
    try:
        url = base_url + str(page)
        response = requests.get(url, headers=headers)
        time.sleep(1)
        soup = BeautifulSoup(response.text, "html.parser")

        repo_blocks = soup.select("li.ListItem-module__listItem--k4eMk")
        print(f"{page} ページ目に {len(repo_blocks)} 個のリポジトリブロックが見つかった")

        for repo in repo_blocks:
            name_tag = repo.select_one("a[data-component='listview-title-link'] span")
            name = name_tag.text.strip() if name_tag else "N/A"

            lang_tag = repo.select_one("span.ReposListItem-module__Text_4--mkG7R")
            language = lang_tag.text.strip() if lang_tag else "N/A"

            star_tag = repo.select_one("a[href$='/stargazers']")
            stars_text = star_tag.text.strip() if star_tag else "0"
            if 'k' in stars_text.lower():
                stars = int(float(stars_text.lower().replace('k',''))*1000)
            else:
                stars = int(stars_text.replace(",", "")) if stars_text else 0

            all_repos.append((name, language, stars))

    except requests.RequestException as e:
        print(f"requests error: {e}")


conn = sqlite3.connect(db_path)
cur = conn.cursor()
cur.execute("""
CREATE TABLE IF NOT EXISTS repos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    language TEXT,
    stars INTEGER
)
""")
conn.commit()


try:
    cur.executemany("INSERT INTO repos (name, language, stars) VALUES (?, ?, ?)", all_repos)
    print(f"{len(all_repos)}line well down!")
except sqlite3.Error as e:
    print(f"db error: {e}")
finally:
    conn.commit()
    conn.close()
```

```python
db_path = "google.db"
conn = sqlite3.connect(db_path)
cur = conn.cursor()

try:
    cur.execute("SELECT * FROM repos")
    rows = cur.fetchall()
    for row in rows:
        id_, name, language, stars = row
        print(f"id: {id_}, name: {name}, language: {language}, stars: {stars}")
except sqlite3.Error as e:
    print(f"db error: {e}")
finally:
    cur.close()
    conn.close()
```
