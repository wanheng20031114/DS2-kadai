## データ　永続化

```python
import sqlite3 
```

```python
path = ""
name = "test.db"

conn = sqlite3.connect(path+name)

conn.close()
```

```python
path = ""
name = "car.db"

conn = sqlite3.connect(path+name)

cur = conn.cursor()

cur.execute("""
CREATE TABLE IF NOT EXISTS cars (
    id INTEGER PRIMARY KEY,
    name TEXT,
    price REAL
)
""")
conn.commit()

cur.close()
conn.close()
```
